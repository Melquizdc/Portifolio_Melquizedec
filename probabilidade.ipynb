```python
# distribuição normal
import scipy.stats as st
def valida_float(pergunta):
  while True:
    try:
      escolha = float(input(pergunta +"\n>> "))
    except ValueError:
      print('')
      continue
    break
  return escolha
def valida_s_n(pergunta):
  while True:
    escolha = input(pergunta +" (S/N) >> ")
    if escolha == "s" or escolha == "n" or escolha == "S" or escolha == "N":
      break
    else:
      print('')
  return escolha

print('calculadora de distribuição normal')
while True:
  try:
    if r != '':
      # ele também considera o desvio padrão do bloco acima
      p = valida_s_n('gostaria de usar as informações do desvio padrão anterior?')
      if p == 's' or p == 'S':
        break
      else:
        r = 'aa'
    else:
      o = valida_float('digite a média')
      desvio = valida_float('digite o desvio padrão')
      r = ''
      break
  except NameError:
    r = 'aa'

x = valida_float('digite o valor desejado')
d = st.norm(o,desvio).cdf(x)-0.5
print(f'a probabilidade é de {"{:.2%}".format(d)}')
```

```python
# amostragem slovin
from math import ceil

def valida_float(pergunta):
  while True:
    try:
      escolha = float(input(pergunta +"\n>> "))
    except ValueError:
      print('')
      continue
    break
  return escolha

print('calculadora de tamanho de amostrage usando formula de slovin')
x = 0
y = 0

x = valida_float('digite o tamanho da população')
y = valida_float('digite a margem de erro em porcentagem')
s = x/(1+x*(y/100)**2)
print(f'o tamanho da amostra é de {ceil(s)}')
```
