```python
# gráficos de funções de até quinto grau
import numpy as np
import matplotlib.pyplot as plt

def valida_int(pergunta):
  while True:
    try:
      escolha = int(input(pergunta +"\n>> "))
    except ValueError:
      print('')
      continue
    break
  return escolha

def valida_float(pergunta):
  while True:
    try:
      escolha = float(input(pergunta +"\n>> "))
    except ValueError:
      print('')
      continue
    break
  return escolha

print('calculadora de gráficos de funções exponenciais até o grau 5')

p = valida_int('qual o grau da função?')
f = valida_float('qual o valor do termo independente')
if p >= 1:
  e = valida_float('qual o valor do termo linear')
if p >= 2:
  d = valida_float('qual o valor do termo quadrático')
if p >= 3:
  c = valida_float('qual o valor do termo cúbico')
if p >= 4:
  b = valida_float('qual o valor do termo quarádico')
if p >= 5:
  a = valida_float('qual o valor do termo quinquênio')

vmin = int(input('digite o valor mínimo do eixo x\n>>'))
vmax = int(input('digite o valor máximo do eixo x\n>>'))
x = np.linspace(vmin,vmax,120)

y = a*x**5+b*x**4+c*x**3+d*x**2+e*x+f
plt.plot(x,y)
plt.show()
```
