```python
# médias e médias ponderadas
import numpy as np

def valida_int(pergunta):
  while True:
    try:
      escolha = int(input(pergunta +"\n>> "))
    except ValueError:
      print('')
      continue
    break
  return escolha
def valida_float(pergunta):
  while True:
    try:
      escolha = float(input(pergunta +"\n>> "))
    except ValueError:
      print('')
      continue
    break
  return escolha

def valida_s_n(pergunta):
  while True:
    escolha = input(pergunta +" (S/N) >> ")
    if escolha == "s" or escolha == "n" or escolha == "S" or escolha == "N":
      break
    else:
      print('')
  return escolha

print('calculadora de média')
q = 0
p = ''
m = []
n = []
o = 0
r = 'aa'

p = valida_s_n('é uma média simples?')

if p == 's' or p == 'S':
  while q < 2:
    q = valida_int('digite a quantidade de termos da média')
  for i in range(1,q+1,1):
    pergunta = valida_float(f'digite o {i}º número')
    m.append(pergunta)
  o = sum(m) / len(m)
  r='aa'
  print(f'a média é {o}')

else:
  q = 0
  while q < 2:
    q = valida_int('digite a quantidade de termos da média')
  for i in range(1,q+1,1):
    pergunta = valida_float(f'digite o {i}º número')
    m.append(pergunta)
    pergunta = valida_float(f'agora, digite o peso do {i}º número')
    n.append(pergunta)

  m = np.array(m)
  n = np.array(n)
  print(np.matmul(m,n))
  o = 'aa'
  r = 'aa'
```

```python
# moda
from collections import Counter

def valida_int(pergunta):
  while True:
    try:
      escolha = int(input(pergunta +"\n>> "))
    except ValueError:
      print('')
      continue
    break
  return escolha
def valida_float(pergunta):
  while True:
    try:
      escolha = float(input(pergunta +"\n>> "))
    except ValueError:
      print('')
      continue
    break
  return escolha

print('calculadora de moda')
x = []
q = 0

mais_comum, frequencia = 0, 0
while q < 3:
  q = valida_int('digite a quantidade de termos da moda')
for i in range(1,q+1,1):
  pergunta = valida_float(f'digite o {i}º número')
  x.append(pergunta)

x = Counter(x)
mais_comum, frequencia = x.most_common(1)[0]
print(f'a moda é {mais_comum} com a frequência de {frequencia}')
```
